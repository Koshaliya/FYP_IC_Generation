```python
import requests
from bs4 import BeautifulSoup
import nltk
import re
import string
import joblib
import pandas as pd
import numpy as np
import seaborn as sns
from nltk.corpus import stopwords
from spacy.lang.ta.stop_words import STOP_WORDS
```

```python
sentance = []
```

```python
# Define the URL of the webpage to scrape
url = "https://www.kidsworldfun.com/ebooks-its-ammas-birthday.php"  # Replace with the desired URL

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find relevant elements containing the sentences or paragraphs
# Use appropriate HTML tags or class attributes to target the desired content
relevant_elements = soup.find_all("p")  # Example: find all <p> elements

# Extract the text content from the relevant elements
sentences = [element.text for element in relevant_elements]

# Filter and keep only the sentences or paragraphs suitable for 6-year-old kids
filtered_sentences = []
for sentence in sentences:
    # Remove any HTML tags or unwanted characters from the sentence
    clean_sentence = re.sub('<.*?>', '', sentence).strip()

    # Check if the sentence is suitable for 6-year-old kids based on certain criteria
    if len(clean_sentence) > 10 and len(clean_sentence.split()) > 3:
        filtered_sentences.append(clean_sentence)

# Print the filtered sentences or do further processing
for sentence in filtered_sentences:
    print(sentence)


for sen in filtered_sentences:
  s = sen.split('.')
  for i in s :
    sentance.append(i)

len(sentance)
```

```python
ssentance = list(dict.fromkeys(sentance))
len(ssentance)
```

```python
import pandas as pd

# Calling DataFrame constructor on list
# with indices and columns specified
df = pd.DataFrame(ssentance, columns =['raw_sentences'])
df
```
